```python
import googlemaps
import pandas as pd
import numpy as np
import os
```

```python
key = ''
gmaps = googlemaps.Client(key=key)
```

```python
df = pd.read_csv('data/coord_train.csv', encoding='euc-kr')
df.iloc[[7]]
```

```python
len(df)
```

```python
address = []
for i in range(len(df)):
    rev_geo = gmaps.reverse_geocode((df.loc[i][1], df.loc[i][2]), language='ko')[0]
    tmp = rev_geo['formatted_address']
    address.append(tmp)
```

```python
address = address.copy()
```

```python
address_df = pd.DataFrame(address, index=None, columns=['지번주소'])
```

```python
train_address = pd.concat([df[['전철역명']], address_df], axis=1)
```

```python
train_address
```

```python
seoul = []
etc = []
for ad in train_address:
    if '서울특별시' in ad.지번주소:
        seoul.append(ad[5:])
    else:
        etc.append(ad[5:])
```

```python
len(etc)
```

```python
miss = []
okay = []
for s in seoul:
    if '동' not in s:
        miss.append(s)
    else:
        okay.append(s)
```

```python
miss_df = pd.DataFrame(miss)
```

```python
miss_df.to_csv('data/train_miss_address.csv', encoding='euc-kr', index=None)
```

```python
train_address.to_csv('data/train_address.csv', encoding='euc-kr', index=False)
```
