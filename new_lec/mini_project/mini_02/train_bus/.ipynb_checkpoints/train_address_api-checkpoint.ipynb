```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import os
import numpy as np
```

```python
def get_address(st_name):
    url = f"http://swopenapi.seoul.go.kr/api/subway/776559696b726b6d3435494b495048/xml/stationInfo/0/7/{st_name}"
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    adr = soup.find_all('adresbass')[-1].text 0
    return adr
```

```python
path = 'data/tmoney_data'
file = os.listdir(path)[0]
file_path = os.path.join(path, file)
file_path
```

```python
df1 = pd.read_excel(file_path, sheet_name='지하철 노선별 역별 이용현황', encoding='euc-kr')
len(df1)
```

```python
def get_stname(x):
    if x.find('(') == -1:
        return x
    else:
        return x[:x.find('(')]
```

```python
ls = df1[["지하철역"]].values
ls = np.reshape(ls, (591))
ls = list(map(get_stname, ls))
```

```python
ls
```

```python
address = []
for l in ls:
    try:
        address.append(get_address(l))
    except IndexError as e:
        print(l, e)
        break
```
