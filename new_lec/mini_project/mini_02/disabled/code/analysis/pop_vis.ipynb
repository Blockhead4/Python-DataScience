```python
data_path = r"D:\workspace-Jwp\python\DataScience\new_lec\mini_project\mini_02\disabled\data"
```

```python
pop = pd.read_csv(os.path.join(data_path, "pop_dis.csv"), encoding='euc-kr')
```

```python
eng_gu = pd.read_csv(os.path.join(data_path, "서울시 행정구역 영어.csv"), encoding='euc-kr')
```

```python
eng_gu
```

```python
pop_gu = pop[pop.동 == '소계']
```

```python
str_to_int = []
for i in range(len(pop_gu.합계)):
    tmp = int(pop_gu.합계.iloc[i].replace(',',''))
    str_to_int.append(tmp)
```

```python
pop_gu.합계 = str_to_int
```

```python
ax = sns.barplot(pop_gu.구, pop_gu.합계)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
vis_df = pop_gu.merge(eng_gu, how='left', left_on='구', right_on='한글')
```

```python
vis_df = vis_df[['한글', '합계']]
```
