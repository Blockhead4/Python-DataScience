### Q1.

두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다. <br>

1. 각각 5 x 5 이상의 크기를 가진다. <br>
2. 공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다. <br>

```python
df_a = pd.DataFrame({
    '이름': ['박진원', '신은총', '권재', '조웅', '른산'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01', '2018-01-03', '2018-01-05'],
    '키': ['180', '170', '170', '174', '173'],
    '몸무게': ['74', '72', '70', '80', '81'],
    '나이': ['32', '31', '32', '31', '32']})
df_a
```

```python
df_b = pd.DataFrame({
    '성명': ['박진원', '신은총', '권재', '조웅', '른산'],
    '투자금': ['100억', '50억', '60억', '40억', '80억'],
    '지분율': ['30%', '15%', '20%', '10%', '25%'],
    '직위': ['대표이사', '사원', '부대표', '반장', '대주주'],
    '배당금': ['20억', '10억', '12억', '8억', '15억']})
df_b
```

```python
pd.merge(df_a, df_b, left_on="이름", right_on="성명")
```

### Q2.

어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 실적 정보는 "매출", "비용", "이익" 으로 이루어진다. (이익 = 매출 - 비용).

또한 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
first_half = pd.DataFrame({
    "날짜(월)" : ["1월", "2월", "3월", "4월", "5월", "6월"],
    "매출" : [100, 300, 500, 1000, 1200, 900],
    "비용" : [50, 100, 200, 400, 500, 300]
})
first_half["이익"] = first_half["매출"] - first_half["비용"]
first_half
```

```python
second_half = pd.DataFrame({
    "날짜(월)" : ["7월", "8월", "9월", "10월", "11월", "12월"],
    "매출" : [1500, 1400, 1700, 2000, 2500, 2100],
    "비용" : [500, 600, 650, 800, 900, 700]
})
second_half["이익"] = first_half["매출"] - first_half["비용"]
second_half
```

```python
overall = pd.concat([first_half, second_half])
overall = overall.set_index("날짜(월)")
overall
```

```python
overall.loc["Total", "매출":] = overall.sum()
overall
```

### Q3.

key1의 값을 기준으로 data1의 값을 분류하여 합계를 구한 결과를 시리즈가 아닌 데이터프레임으로 구한다.

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
df2.groupby(df2.key1).sum()[["data1"]]
```

### Q6.

타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.

```python
titanic = sns.load_dataset("titanic")
titanic.tail()
```

1. qcut 명령으로 세 개의 나이 그룹을 만든다.

```python
titanic["age_group"] = pd.qcut(titanic.age, 3, labels=["1", "2", "3"])
titanic.tail()
```

2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.

```python
def prob_survived(x):
    return x.sum() / x.count()
```

```python
titanic.pivot_table('survived', ['sex', 'age_group'], 'class', aggfunc=prob_survived)
```

3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
titanic.pivot_table('survived', 'sex', 'class', aggfunc=prob_survived).head()
```
