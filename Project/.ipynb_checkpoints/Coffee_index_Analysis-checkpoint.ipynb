## 커피지수 분석

각 지역별 커피지수를 구하여 시각화 한다.
- Starbucks, Coffee Bean, Ediya, 빽다방 커피 매장수를 파악하여 커피지수를 계산한다.
- 커피지수 = (Starbucks + Coffee Bean + Ediya) / 빽다방

### 1. 데이터 크롤링

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

- Starbucks Crawling

```python
data = {
    'ins_lat':'', # 지정한 위도와 경도에서 가까운 순으로 나열
    'ins_lng':'',
    'p_sido_cd':'', # 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd':'',  # 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd':'',
    'set_date':'',
    'iend':'1000000',
}
    
url = 'https://www.istarbucks.co.kr/store/getStore.do'
r = requests.post(url, data=data)

r.text[:1000] # 수신된 데이터의 앞부분만 확인
```

```python
jo = json.loads(r.text)
star_df = json_normalize(jo, 'list')
```

- 행(row)수, 전국 457개 매장

```python
len(star_df)
```

```python
# 컬럼수 111개
star_df.columns
```

- 필요한 컬럼만 선택

```python
# df = df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address', 'tel']]
star_df = star_df[['s_name', 'sido_name', 'gugun_name']]
star_df.head()
```

```python
star_df2 = star_df.copy()
star_df2 = star_df2.groupby([star_df2.sido_name, star_df2.gugun_name])
star_df2 = star_df2.count()
star_df2.columns = ['starbucks']
star_df2
```

- Coffee Bean Crawling

```python
sido = ['서울', '경기', '인천', '강원', '경남', '경북', '광주', '대구', '대전', '부산', '울산', '충남', '충북']

raw = ''
for v in sido:
    url = 'https://www.coffeebeankorea.com/store/store_data2.asp?lat=&lng=&chk1=0&chk2=0 \
           &chk3=0&chk4=0&chk5=0&chk6=0&chk7=0&chk8=0&chk9=0&keyword=&StoreLocal=%s' % v
    r = requests.post(url)
    
    raw += r.text
```

```python
raw2 = raw.split()
raw2
```

```python
coffee_bean = []
for i in range(len(raw2)-3):
    if raw2[i] == 'StoreAddress':
        coffee_bean.append([raw2[i+2], raw2[i+3]])
```

```python
coffee_bean[3] = ["'서울특별시", "송파구"]
```

```python
for i in range(len(coffee_bean)):
    coffee_bean[i][0] = coffee_bean[i][0].replace("'", '')
```

```python
cb_df = pd.DataFrame(coffee_bean)
cb_df.columns = ['sido_name', 'gugun_name']
cb_df['coffe_bean'] = 1
cb_df = cb_df.groupby(['sido_name', 'gugun_name'])
cb_df = cb_df.count()
cb_df
```

import requests
from urllib import request, parse
import time
from bs4 import BeautifulSoup
from multiprocessing import Pool
import socket
import os

# Set the default timeout in seconds
timeout = 5
socket.setdefaulttimeout(timeout)

# avoid forbidden
opener = request.build_opener()
opener.addheaders = [('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
request.install_opener(opener)

base_url = 'http://pixabay.com/videos/'

def crawl():
    video_links = []
    n = 0
    for i in range(76):
        i+=1

        url = base_url + f'search/?pagi={i}&resolution_hd=1'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")

        for item in soup.findAll("div",{'class':'item'}):
            n+=1
            key = item.find("a").get('href').replace('/videos/','').replace('/','')
            name = key+'.mp4'
            link = parse.urljoin(base_url, '/'.join(['download', key+'_large.mp4?attachment']))
            video_links.append([n, name, link])
        print(f'page {n} crawl complete')
    return video_links

def download(video_links):
    i = video_links[0]
    name = video_links[1]
    link = video_links[2]
    if not os.path.isfile(name):
        try:
            request.urlretrieve(link, name)
            print(f'Download complete. {i} : {name}')
        except:
            print(f'Download failed. {i} : {name}')
    else:
        print(f'Already exist. {i} : {name}')
    return


if __name__ == "__main__":
    video_links = crawl()
    pool = Pool(processes=4)
    pool.map(download, video_links)
