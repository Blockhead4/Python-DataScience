### Q1.

투수들의 기록중에서 평균자책점(ERA), 투구인닝(IP), 탈삼진(SO) 기록을 찾아서 Pitcher_stats 란 테이블을 만들고, Eagles 테이블과 Join 하여 백넘버, 선수명, 포지션, 투구인닝, 평균자책점, 탈삼진 필드를 갖는 데이터 프레임을 만들어서 Join 한 결과를 입력하고, 그 결과를 보이시오.

```python
cur = conn.cursor()
cur.execute('CREATE TABLE IF NOT EXISTS Pitcher_Stats \
            (id INT NOT NULL, \
             name TEXT, \
             era REAL, \
             ip REAL, \
             so INT, \
             PRIMARY KEY(id));')
```

```python
pitcher_stats = pd.read_csv('pitcher_stats.csv', encoding='EUC_KR')
pitcher_stats
```

```python
cur = conn.cursor()
sql = "INSERT INTO Pitcher_Stats VALUES (?, ?, ?, ?, ?);"
for i in range(1, 26):
    cur.execute(sql, (i, *pitcher_stats.values[i-1]))
conn.commit()
```

```python
cur = conn.cursor()
cur.execute('SELECT * FROM Pitcher_Stats')
cur.fetchall()
```

```python
sql = "SELECT e.back_no, e.name, e.position, \
              s.era, s.ip, s.so \
              FROM Eagles AS e JOIN Pitcher_Stats AS s \
              ON e.name LIKE s.name;"
cur = conn.cursor()
cur.execute(sql)
join = cur.fetchall()
join
```

```python
column_name = ['등번호', '선수명', '포지션', '방어율', '투구인닝', '탈삼진']
pitcher_df = pd.DataFrame(join, columns = column_name)
pitcher_df
```

### Q3.

사용자의 이름과, 비밀번호를 갖는 Users 테이블이 있다. 사용자의 이름과 비밀번호를 올바르게 입력하면 ‘성공’을 출력하고, 잘못 입력하면 ‘실패’를 출력하는 프로그램을 작성하시오.

```python
cur = conn.cursor()
cur.execute('CREATE TABLE IF NOT EXISTS Users \
            (id INT NOT NULL, \
             user_name TEXT, \
             password TEXT, \
             PRIMARY KEY(user_name));')
```

```python
user = pd.read_csv('user.csv', encoding='euc-kr')
user
```

```python
cur = conn.cursor()
sql = "INSERT INTO Users VALUES (?, ?, ?);"
for i in range(len(user)):
    cur.execute(sql, user.values[i])
conn.commit()
```

```python
cur = conn.cursor()
cur.execute('SELECT * FROM Users')
cur.fetchall()
```

```python
def login(user_name, password):
    sql = f"SELECT user_name, password FROM Users WHERE user_name = '{user_name}' AND password = '{password}'"
    cur = conn.cursor()
    cur.execute(sql)
    return print('성공') if cur.fetchall() else print('실패')
```
